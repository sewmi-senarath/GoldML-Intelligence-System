```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('../../data/Gold_Market_10y.csv')
df.head()
df.info()
df.describe()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
x = df[['Open','High','Low','Volume','MA_20','MA_50','Volatility_20']]
y = df['Close']
```

```python
df = df.reset_index(drop=True)
df.info()
df.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y, test_size=0.2, random_state=42
)
```

```python
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = SVR(kernel='rbf')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("MSE: ", mean_squared_error(y_test, y_pred)) 
# Measures prediction error

print("R2: ", r2_score(y_test,y_pred))
# Model explains ~79% of variance in gold prices. 0.8+ → very good
```

```python
x = df[['Volume', 'Year', 'Month', 'Day_of_Week', 'Quarter']]
y = df['Close']
```

```python
# Re-split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Re-scale the data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Re-train the model
model = SVR(kernel='rbf', C=100, gamma=0.1)
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)
print("MSE: ", mean_squared_error(y_test, y_pred))
print("R2: ", r2_score(y_test, y_pred))
```

```python
results = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

results.to_csv("svr_results.csv", index=False)
```

```python
plt.figure()
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("SVR: Actual vs Predicted")
plt.show()
```

```python
plt.figure()

plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')

plt.legend()
plt.title("SVR Prediction vs Actual")
plt.show()
```

```python
residuals = y_test - y_pred

plt.figure()
plt.scatter(y_pred, residuals)
plt.axhline(y=0)
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

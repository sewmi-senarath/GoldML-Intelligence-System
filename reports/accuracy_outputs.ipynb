```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Linear\nRegression', 'Random Forest\nRegressor', 'Gradient Boosting\nRegressor', 'Support Vector\nRegressor']
mae    = [2.6042, 2.698227, 0.9281, 0.4186]
rmse   = [3.7513, 3.822797, 1.3013, 3.5185]
r2     = [0.9964, 0.995997, 0.9974, 0.9971]

x = np.arange(len(models))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))

bars1 = ax.bar(x - width, mae,  width, label='MAE',       color='steelblue')
bars2 = ax.bar(x,         rmse, width, label='RMSE',      color='darkorange')
bars3 = ax.bar(x + width, r2,   width, label='R² Score',  color='seagreen')

ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Model Performance Comparison: MAE, RMSE, and R² Score', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=10)
ax.legend(fontsize=11)
ax.grid(axis='y', linestyle='--', alpha=0.5)

# Add value labels on top of each bar
for bar in [bars1, bars2, bars3]:
    for b in bar:
        ax.annotate(f'{b.get_height():.4f}',
                    xy=(b.get_x() + b.get_width() / 2, b.get_height()),
                    xytext=(0, 4), textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('model_comparison_chart.png', dpi=150)
plt.show()
```
